```python
# Check for missing values
print(df.isnull().sum())

# Numaratic col name
numeric_columns = df.select_dtypes(include=[np.number]).columns

# ortalama değir ile doldurup hata almasını engelemek amçlandı
df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].mean())

# Kategorik sütunlardaki eksik değerleri en sık görülen değer (mod) ile doldur
categorical_columns = df.select_dtypes(include=[object]).columns
for column in categorical_columns:
    df[column].fillna(df[column].mode()[0], inplace=True)

# Eksik değerlerin kontrolünden sonra tekrar kontrol et
print(df.isnull().sum())
```

```python
# Handling missing values
for column in df.select_dtypes(include=[np.number]).columns:
    df[column].fillna(df[column].mean(), inplace=True)
    
```

```python
# Outlier detection and treatment using the IQR method
def treat_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[column] = np.where(df[column] < lower_bound, lower_bound, df[column])
    df[column] = np.where(df[column] > upper_bound, upper_bound, df[column])

# Apply outlier treatment for all numerical columns
for column in df.select_dtypes(include=[np.number]).columns:
    treat_outliers(df, column)

# Display the first few rows of the cleaned dataset
print(df.head())

# Display dataset information to ensure missing values are handled and data types are correct
print(df.info())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier, RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

```python
# Verilerrr
numeric = ['age', 'impluse', 'pressurehight', 'pressurelow', 'glucose', 'kcm', 'troponin']

# Veri çerçevesinin gender class için dumy değer
hd_data = pd.get_dummies(df, columns=['gender', 'class'])

# Veri kümesinden (data3) özelliklerin (X) ve etiketlerin (y) ayrılması
X = df.drop('class', axis=1)  # X, class sütunu hariç tüm sütunlar
y = df['class']  # class stunu
```

```python
def relatory_ml(df, algorit, opt=2):
    # Verilerin xy ayarı
    X = df.drop('class', axis=1) 
    y = df['class']
    
    # Verilerin eğitim ve test setlerine bölünmesi
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
    
    # Belirtilen seçeneğe göre model oluşturulması
    if opt == 0:
        ml = algorit(max_iter=1000)
    elif opt == 1:
        ml = algorit(n_estimators=1000)
    elif opt == 2:
        ml = algorit()
    
    # Modelin eğitilmesi
    ml.fit(X_train, y_train)
    
    # Eğitim doğruluğunun hesaplanması ve gösterilmesi
    score_train = ml.score(X_train, y_train)
    print(f"Eğitim Doğruluğu: {score_train * 100:.2f}%")
    
    # Test doğruluğunun hesaplanması ve gösterilmesi
    score_test = ml.score(X_test, y_test)
    print(f"Test Doğruluğu: {score_test * 100:.2f}%")
    
    # Test setindeki etiketlerin tahmin edilmesi
    y_predicted = ml.predict(X_test)
    
    # Sınıflandırma raporunun gösterilmesi
    print("------- SINIFLANDIRMA RAPORU -------")
    print(classification_report(y_test, y_predicted))
    
    # Karışıklık matrisinin gösterilmesi
    print("------- KARIŞIKLIK MATRİSİ -------")
    confusion = confusion_matrix(y_test, y_predicted)
    sns.heatmap(confusion, annot=True, cmap="Purples")
    
    # Eğitim ve test doğruluklarının döndürülmesi
    return score_train, score_test
```

```python
# LogisticRegression modeliyle eğitim ve test doğruluklarının hesaplanması
log_train, log_test = relatory_ml(df, LogisticRegression, 0)
```

```python
# DecisionTreeClassifier modeliyle eğitim ve test doğruluklarının hesaplanması
tree_train, tree_test = relatory_ml(df, DecisionTreeClassifier)
```

```python
# AdaBoostClassifier modeliyle eğitim ve test doğruluklarının hesaplanması
ada_train, ada_test = relatory_ml(df, AdaBoostClassifier)
```

```python
# KNeighborsClassifier modeliyle eğitim ve test doğruluklarının hesaplanması
kn_train, kn_test = relatory_ml(df, KNeighborsClassifier)
```

```python
# RandomForestClassifier modeliyle eğitim ve test doğruluklarının hesaplanması
rand_train, rand_test = relatory_ml(df, RandomForestClassifier)
```

```python
# GradientBoostingClassifier modeliyle eğitim ve test doğruluklarının hesaplanması
gboost_train, gboost_test = relatory_ml(df, GradientBoostingClassifier)
```

```python
# GradientBoostingClassifier modeliyle eğitim ve test doğruluklarının hesaplanması
gboost_train, gboost_test = relatory_ml(df, GradientBoostingClassifier)
```
