```python
import pandas as pd 
import seaborn  as sns
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# File Reade
file_path = r'C:\Users\Tan\Desktop\Seng352l\Heart Attack.csv'
df = pd.read_csv(file_path)
```

```python
# Show data
print(df)
```

```python
# Compute the mean age
mean_age = df['age'].mean()
print("Mean Age:", mean_age)

# Compute the standard deviation of age
std_dev_age = df['age'].std()
print("Standard Deviation of Age:", std_dev_age)

# Find the minimum and maximum age
min_age = df['age'].min()
max_age = df['age'].max()
print("Minimum Age:", min_age)
print("Maximum Age:", max_age)

# Calculate the 25th, 50th (median), and 75th percentiles of age
q1_age = df['age'].quantile(0.25)
median_age = df['age'].median()
q3_age = df['age'].quantile(0.75)

print("25th Percentile of Age:", q1_age)
print("Median Age:", median_age)
print("75th Percentile of Age:", q3_age)
```

```python
#Data Cleaning and Preprocessing
#Missing values
print(df.isnull().sum())
#Check for Duplicates
print("Number of dup:",df.duplicated().sum())
```

```python
#Keşifsel Veri Analizi (Exploratory Data Analysis - EDA)
df.hist(bins = 30,figsize =(20,10))
plt.show()
```

```python
# Numerik sütunları seçme
numeric_columns = df[['age', 'gender', 'impluse', 'pressurehight', 'pressurelow', 'glucose', 'kcm', 'troponin']]

# Korelasyon matrisini hesaplama
corr_matrix = numeric_columns.corr()

# Korelasyon matrisini görselleştirme
plt.figure(figsize=(12,10))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.show()
```

```python
# age:who had a heart attack
fig, axes = plt.subplots(1, 3, figsize=(13, 5))
plt.suptitle('Age : Heart Attack', fontsize=25, fontweight='bold', color='navy')
fig.tight_layout(pad=2.0)
# ax1
sns.scatterplot(data=df, x='age', y='class', ax=axes[0], hue='class')

# ax2
sns.histplot(data=df, x='age', ax=axes[1])

# ax3
sns.boxplot(data=df, x='age', y='class', ax=axes[2])

plt.show()
```

```python
#  impluse:who had a heart attack
fig, axes = plt.subplots(1, 3, figsize=(13, 5))
plt.suptitle('İmpluse : Heart Attack', fontsize=25, fontweight='bold', color='navy')

# ax1
sns.scatterplot(data=df, x='impluse', y='class', ax=axes[0], hue='class')

# ax2
sns.histplot(data=df, x='impluse', ax=axes[1])

# ax3
sns.boxplot(data=df, x='impluse', y='class', ax=axes[2])

plt.show()

df[df.impluse>1000]
```

```python
# Gender : Heart Attack
fig, axes = plt.subplots(1, 3, figsize=(13, 5))
plt.suptitle('Gender : Heart Attack', fontsize=25, fontweight='bold', color='navy')
fig.tight_layout(pad=2.0)

# ax1
sns.scatterplot(data=df, x='gender', y='class', ax=axes[0], hue='class')
axes[0].set_xticks([0, 1])
axes[0].set_xticklabels(np.array(['Male', 'Female']))

# ax2
sns.histplot(data=df, x='gender', ax=axes[1])
axes[1].set_xticks([1, 1])
axes[1].set_xticklabels(np.array(['Male', 'Female']))

# ax3
sns.boxplot(data=df, x='gender', y='class', ax=axes[2])
axes[2].set_xticks([0, 1])
axes[2].set_xticklabels(np.array(['Male', 'Female']))

plt.show()
```

```python
# Target
fig, axes = plt.subplots(1, 2, figsize=(15, 4))
# ax1
d = df[['class']]
d = d.groupby('class').size()
labels = ['Negative', 'Positive']
explode = (0, 0.1)
axes[0].pie(d, labels=labels, explode=explode, autopct='%1.1f%%', shadow=True)

# ax 2
c = ['red', 'blue']
x = df['class'].value_counts().sort_values().index
y = df['class'].value_counts().sort_values()
bar_container = axes[1].bar(x, y, color=c, width=1, edgecolor="white", linewidth=0.9)
axes[1].bar_label(bar_container)
axes[1].set(xlabel='classes', ylabel='abundance', xticks=np.arange(0, 1.5))
plt.show()
```
